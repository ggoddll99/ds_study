### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips=pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv", engine='python')
tips
```

### Step 4. Delete the Unnamed 0 column

출처: https://benn.tistory.com/27 [밴의 데이터 과학:티스토리]

```python
tips = tips.drop(tips.columns[[0]], axis=1)
tips
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(tips["total_bill"])
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill")
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
tips.plot(kind='scatter', x='total_bill', y='tip')
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title('Scatter Plot - total_bill vs tip')

plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
tips.describe()
```

```python
# 3개의 변수로 산점도 그리기
tips.plot(kind='scatter', x='total_bill', y='tip', c='size')

plt.title('Scatter Plot : total_bill-tip-size')

plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Relationship between Days and Total Bill")

plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(x="tip", y="day", hue="sex", data=tips)
plt.xlabel("Tip")
plt.ylabel("Day")
plt.title("Tip vs Day (Differentiated by Sex)")
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x="day", y="total_bill", hue='time', data=tips)
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Relationship between Days and Total Bill and Time")

plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
# 그래프 객체 생성(figure에 2개의 서브 플롯 생성)
fig=plt.figure()
ax1=fig.add_subplot(1, 2, 1)
ax2=fig.add_subplot(1, 2, 2)

# 그래프 그리기
sns.histplot(x=tips[tips['time']=='Dinner']['tip'], ax=ax1)
sns.histplot(x=tips[tips['time']=='Lunch']['tip'], ax=ax2)

plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
# 그래프 객체 생성(figure에 2개의 서브 플롯 생성)
fig=plt.figure()
ax1=fig.add_subplot(1, 2, 1)
ax2=fig.add_subplot(1, 2, 2)

# 그래프 그리기
sns.scatterplot(x=tips[tips['sex']=='Male']['total_bill'], y=tips[tips['sex']=='Male']['tip'], hue=tips[tips['sex']=='Male']['smoker'], ax=ax1)
sns.scatterplot(x=tips[tips['sex']=='Female']['total_bill'], y=tips[tips['sex']=='Female']['tip'], hue=tips[tips['sex']=='Female']['smoker'], ax=ax2)

plt.show()
```
