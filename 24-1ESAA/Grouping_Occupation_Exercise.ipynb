### Step 1. Import the necessary libraries

### Step 3. Assign it to a variable called users.



```python
users=pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep='|', engine='python')
users
```

### Step 4. Discover what is the mean age per occupation



```python
users.groupby('occupation').mean()['age'] # occupation 열을 기준으로 그룹화, 합계 연산
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users['is_male']=(users['gender']=='M')
users
```

```python
(users.groupby('occupation')['is_male'].sum() / users.groupby('occupation')['gender'].count()).sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
print(users.groupby('occupation')['age'].max())
print(users.groupby('occupation')['age'].min())
```

* 한 번에 표로 정리

참고 : https://www.kaggle.com/code/daguangya/pandas-exercises-3-2

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).mean()[['age']]
```

### Step 8.  For each occupation present the percentage of women and men

```python
(users.groupby(['occupation', 'gender']).agg({'gender': 'count'}) / users.groupby(['occupation']).count())[['gender']]
```
